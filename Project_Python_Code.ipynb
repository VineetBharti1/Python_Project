```python
data_list = []
add_header(data_list)
# print(data_list)
arr = np.array(data_list)
for k,v in data.items():
    data_list_row = []
    data_list_row.append(k)
    try:
        data_list_row.append(v["delta"]["confirmed"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["delta"]["tested"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["delta"]["deceased"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["delta"]["recovered"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["delta"]["vaccinated1"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["delta"]["vaccinated2"])
    except KeyError:
        data_list_row.append("NaN")
            
    
    try:
        data_list_row.append(v["delta21_14"]["confirmed"])
    except KeyError:
        data_list_row.append("NaN")
        
     
    try:
        data_list_row.append(v["delta7"]["confirmed"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["delta7"]["tested"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["delta7"]["deceased"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["delta7"]["recovered"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["delta7"]["vaccinated1"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["delta7"]["vaccinated2"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["meta"]["population"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["total"]["confirmed"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["total"]["tested"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["total"]["deceased"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["total"]["recovered"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["total"]["vaccinated1"])
    except KeyError:
        data_list_row.append("NaN")
    try:
        data_list_row.append(v["total"]["vaccinated2"])
    except KeyError:
        data_list_row.append("NaN")
    row = np.array(data_list_row)
    # print(row)
    arr = np.vstack((arr, row))                           
    # print(arr)
```

```python
with open("State_Covid_19.csv", "w") as f:
    writer = csv.writer(f)
    writer.writerows(arr)
f.close()
```

```python
r=requests.get(url="https://data.covid19india.org/v4/min/timeseries.min.json")
```

```python
x=pd.DataFrame.transpose(pd.DataFrame(data))
```

```python
x.insert(0,"State",np.array(x.index))
```

```python
diii={}
for i in range(int(len(x))):
    value=list(dict(np.array(x)[i][1]).keys())
    key=(np.array(x)[i][0])
    diii[key]=value
```

```python
stat=[]
dates=[]
for i,j in diii.items():
    for k in range(len(j)):
        stat.append(i)
        dates.append(j[k])
```

```python
dis=pd.Series(dates)
st=pd.Series(stat)
l={}
State_dates=pd.DataFrame(l)
```

```python
State_dates.insert(0,"State",st)
State_dates.insert(1,"Date",dis)
#State_dates
```

```python
Concat_values=pd.DataFrame(x.loc[:,"dates"][0]).transpose()
for i in range(1,len(x.loc[:,"dates"])):
    first_st=pd.DataFrame(x.loc[:,"dates"][i]).transpose()
    Concat_values=pd.concat([Concat_values,first_st])
```

```python
col1 = np.array(Concat_values.index)
Concat_values.insert(0,"Date",col1)
Concat_values
```

```python
def to_dataframe(column1,column2):
    vv=np.array(Concat_values.loc[:,[column1,column2]])
    DI={}
    for i in range(len(vv)):
        DI[vv[i][0]]=vv[i][1]
    column2=pd.DataFrame(DI).transpose()
    co=np.array(column2.index)
    column2.insert(0,"Date",co)
    column2.reset_index(drop=True, inplace=True)
    return column2
```

```python
n=to_dataframe(column1="Date",column2="delta")
```
